```python

網頁由三個東西構成:
1. 內容(main): HTML
2. 排版: CSS
3. 動態: JavaScript語法
!!! 某個動作觸發了某些程式碼， 最後送出一個網址去得到更多東西
!!! 但我們care的是那個網址， 很少很少很少你會真的想知道他怎麼觸發的
!!! 所以動態對你而言不需要認真研究

```

```python

CSS語法: 選到你要排版的區塊， 加上排版
1. 利用區塊的名子:
a{
  名:值
}
!!! 但大部分時間我不是要選所有的超連結
!!! 以我們上面介紹的HTML， 我是分不出來特定的超連結
!!! 所以我們會在HTML區塊上， 加上特別的屬性
!!! 這特別的屬性是為了排版而生，讓我可以分出這些區塊
!!! 兩個特別屬性
!!! class="分類1 分類2 分類3"
2. 利用分類(class)屬性
(e.g. class="good bad")
.good{
}

!!! id="身分證"
3. 針對id屬性
(e.g. id="abc")
abc{
}

a{
    background-color:red;
}

# 10px 20px
.js-thumbnail-img {
    border-radius: 50%;
}


```
