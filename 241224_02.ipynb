````python
# bmi = weight(kg) / hight(m) ^ 2
weight = float(input("請輸入體重:"))
hight = float(input("請輸入身高:"))
bmi = weight / (hight / 100) ** 2
print("你的體重是:", weight)
print("你的身高是:", hight)
print("BMI計算為:", bmi)

if bmi > 25 :
  print("你的體重過重")
elif bmi < 18.5:
  print("你的體重過輕")
else:
  print("你的體重在正常範圍")
````

# 作業
6種拳，每三人分一組


image.png

````python
import random

my = int(input("0-5:"))
com = random.randint(0, 5)
# my = 1
# com = 5
trans = ["剪刀(0)", "石頭(0)", "布(0)",
      "剪刀(1)", "石頭(1)", "布(1)"]
print("我出的:", trans[my])
print("電腦的:", trans[com])

# 第一步:先確定他們是否同組
# 0 // 3: 0
# 1 // 3: 0
# 2 // 3: 0
# 3 // 3: 1
# 4 // 3: 1
# 5 // 3: 1
# 6 // 3: 2
# 取商: 分組 取餘: 循環
mygroup = my // 3
comgroup = com // 3
if mygroup == comgroup:
  # 情況1: my, com都是0~2(0~2)
  # 情況2: my, com都是3~5(0~2)
  # 如果是3~5 回到0~2(餘數)
  my2 = my % 3
  com2 = com % 3
  #
  if my2 == (com2 + 1) % 3:
    print("WIN")
  elif com2 == (my2 + 1) % 3:
    print("LOSE")
  else:
    print("平手")
else:
    print("平手(不同組)")

# 當你解決不了問題的時候，先把問題簡化成簡單的每一步驟
````
