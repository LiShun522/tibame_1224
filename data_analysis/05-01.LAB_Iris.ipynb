```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
data = pd.read_csv("../data/iris.csv")
data.head(10)
```

```python
# 繪製散佈圖
sns.scatterplot(data=data, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
# 設定標題和標籤
plt.title('Petal Length vs Petal Width by Species', fontsize=16)
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)
plt.show()
```

```python
# 篩選三種物種的數據
setosa = data[data['Species'] == 'Iris-setosa']
versicolor = data[data['Species'] == 'Iris-versicolor']
virginica = data[data['Species'] == 'Iris-virginica']

# 創建圖表
plt.figure(figsize=(8, 6))

# 繪製散佈圖
plt.scatter(setosa['PetalLengthCm'], setosa['PetalWidthCm'], color='blue', label='Setosa', alpha=0.8)
plt.scatter(versicolor['PetalLengthCm'], versicolor['PetalWidthCm'], color='orange', label='Versicolor', alpha=0.8)
plt.scatter(virginica['PetalLengthCm'], virginica['PetalWidthCm'], color='green', label='Virginica', alpha=0.8)

# 添加標題與標籤
plt.title('Petal Length vs Petal Width by Species', fontsize=16)
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)

# 顯示圖例
plt.legend(title='Species')
plt.grid(alpha=0.3)
plt.show()
```
