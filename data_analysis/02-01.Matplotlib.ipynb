## 使用Matplot來製作圖表

## 線條顏色

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.7')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
# https://www.w3schools.com/tags/ref_colornames.asp
```

## 線條樣式

```python
# 參數alpha可以設定直方圖的透明度，用在多個直方圖重疊時
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.hist(x1, alpha=0.3)
plt.hist(x2, alpha=0.3)
plt.hist(x3, alpha=0.3)
```

```python
# 練習題
# 請畫出隨便一個feature的直方圖，且需要按照不同類別來分色
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
target = iris.target
x1 = features[2][target == 0]
x2 = features[2][target == 1]
x3 = features[2][target == 2]
plt.hist(x1, alpha=0.3, color="r")
plt.hist(x2, alpha=0.3, color="b")
plt.hist(x3, alpha=0.3, color="g")
plt.show()
```

```python
# 設定圖例的位置
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal') # 等比形式
plt.legend(loc='upper right') # 標誌在右上
```

```python
# 練習題
# 請將上圖的圖例放在左下角位置
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal') # 等比形式
plt.legend(loc='lower right') # 標誌在右上
```

```python
# 圖例加框
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend(loc='upper right', frameon=False) # 標示圖要不要有框框，預設是True
```

```python
# 圖例欄數
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.plot(x, np.cos(x+2), ':r', label='cos(x)+2')
plt.axis('equal')
plt.legend(loc='lower center', frameon=True, ncol=2) # 圖示中每行幾個值
```

```python
# 練習題
# 請用 2x2 的圖形分別畫出 sin cos tan tanh 四個圖
# 先劃出格子 2x2

# for i in range(1,5):
plt.subplot(2, 2, 1) # 2x2從第一張圖開始
# 計算出不同值
plt.plot(x, np.sin(x), '-g', label='sin(x)')         # 計算值 顏色
plt.legend(loc='lower center', frameon=True, ncol=2) # 加入標記
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), '-r', label='cos(x)')
plt.legend(loc='lower center', frameon=True, ncol=2)
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x), '-b', label='tan(x)')
plt.legend(loc='lower center', frameon=True, ncol=2)
plt.subplot(2, 2, 4)
plt.plot(x, np.tanh(x), '-d', label='tanh(x)')
plt.legend(loc='lower center', frameon=True, ncol=2)
```

### 使用不等大小子圖的技巧

```python
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up the axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2) # ok = 畫點 、 黑色

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray') 
x_hist.invert_yaxis()  # 垂直顛倒

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis() # 垂直顛倒
```
