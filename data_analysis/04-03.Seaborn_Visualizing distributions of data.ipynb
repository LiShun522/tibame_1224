# Normalized histogram statistics
 stat 

```python
fig, axes = plt.subplots(1, 4, figsize=(25, 8))
#displot()subplots()histplot()
sns.histplot(penguins, x="flipper_length_mm", hue="species", ax=axes[0]) 

# stat 
sns.histplot(penguins, x="flipper_length_mm", hue="species", stat="density", ax=axes[1]) # ""

# common_norm=False
sns.histplot(penguins, x="flipper_length_mm", hue="species", stat="density", common_norm=False, ax=axes[2])

sns.histplot(penguins, x="flipper_length_mm", hue="species", stat="probability", ax=axes[3])
```

# (Distribution visualization in other settings)

### (Plotting joint and marginal distributions)

 jointplot()jointplot()  scatterplot()  histplot() 

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

 displot()  jointplot()  kind="kde"  kdeplot() 

```python
sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde"
)
```

jointplot()  JointGrid 

```python
g = sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm") # 
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
```

### (Plotting many distributions) !!!  !!!
pairplot() (joint distributions)(marginal distributions)pairplot() (small-multiple)

```python
sns.pairplot(penguins) #  !!!
sns.pairplot(penguins, hue = "species")
```

```python
#
#
iris = sns.load_dataset("iris")
iris.head()
```

 jointplot()/JointGrid  PairGrid 

```python
g = sns.PairGrid(penguins, hue = "species")
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
