```python
data.groupby('Platform')[['NA_Sales','Global_Sales']].aggregate(
    [lambda x: x.max() - x.min(),
    lambda x: x.median() - x.mean()]
)
```

```python
# 定義了兩個func，
```

```python
def A1(x):
    return np.max(x) - np.min(x)
def B2(x):
    return np.median(x) - np.mean(x)
# 自定義的函式，但實務上很少用，因為別人不知道你寫甚麼意思 ，所以會用上列大家適用的
data.groupby('Platform')[['NA_Sales','Global_Sales']].aggregate(
    [A1,B2]
)

# x = np.array([1,2,3,4,10])
# A1(x),B2(x)

# lamda 有時候 只需要一個一次性func，但閱讀性較差，寫多人程式適用，不要太多 def，怕名稱打架
# data.groupby('Platform')[['NA_Sales','Global_Sales']].aggregate(
#     [lambda x: x.max() - x.min(),     # 對應 def A1(x)
#     lambda x: x.median() - x.mean()]  # 對應 def B2(x)
# )
```

## Transform

```python
data.groupby('Platform')['Global_Sales'].sum()
```

```python
# 將新的值塞到df，且每個對應
x = data.groupby('Platform')['Global_Sales'].transform(sum) # 每筆資料都要對英的資料
data["各平台銷售總合"] = x
data
```

```python
data.groupby('Platform')['Global_Sales'].aggregate(np.sum)
```

## apply

```python
# 不使用 apply 的寫法
data2 = data.copy()
global_sum = data2.groupby('Platform')['Global_Sales'].transform(np.sum) # 每筆資料都要對英的資料
data2['銷售佔比'] = data2['Global_Sales'] / global_sum
data2
```

```python
#使用apply來達成等同上面的功能
def rate(x):
    x['銷售佔比'] = x['Global_Sales'] / x['Global_Sales'].sum()
    return x
data2 = data.copy()
data2.groupby('Platform').apply(rate) # 經常使用可以用
```

## filter

```python
# 列出該平台總銷售額大於 1000 百萬的遊戲
index = data.groupby('Platform')['Global_Sales'].transform(sum) > 1000 # 實際上比較會這樣用
data[index]
```

```python
# 列出所有PS平台的遊戲
def AAA(x):
    slist = ['PS','PS2','PS3']
    return any(x.iloc[0]['Platform'] in s for s in slist)
data.groupby('Platform').filter(AAA) # 過濾器 過濾你寫的或某些函式
```
