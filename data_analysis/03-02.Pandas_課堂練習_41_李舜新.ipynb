```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("../data/vgsales.csv")
data.head()
```

```python
# 找出有缺失值的欄位
pd.DataFrame(data.isnull().sum())
```

```python
# 計算欄位的基本統計數據
data.describe()
```

```python
# 取出 "Publisher" 欄位的資料
pd.DataFrame(data["Publisher"])
```

```python
# 取出 Publisher, Name, Genre 欄位的資料
pd.DataFrame(data[["Publisher", "Name", "Genre"]])
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和
pd.DataFrame(data[["NA_Sales","EU_Sales","JP_Sales","Other_Sales"]].sum(axis=1) )# 沒設定 axis=1 是行
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和，並存入新的欄位"總和"
data["總和"] = data[["NA_Sales","EU_Sales","JP_Sales","Other_Sales"]].sum(axis=1) # 沒設定 axis=1 是行
pd.DataFrame(data["總和"])
```

```python
#找出發行商為Nintendo的遊戲名稱
pd.DataFrame(data[data["Publisher"] == "Nintendo"]["Name"])
```

```python
#找出發行商為Nintendo的遊戲名稱 且列出不重複的
pd.DataFrame(data[data["Publisher"] == "Nintendo"]["Name"].unique())
data[data["Publisher"] == "Nintendo"]["Name"].unique().size
```

```python
#列出所有Wii的遊戲名稱
pd.DataFrame(data[data["Platform"] == "Wii"]["Name"])
```

```python
#計算所有GBA遊戲的銷售總合
print(pd.DataFrame(data[data["Platform"] == "GBA"]["Global_Sales"]))
print("GBA銷售總和: ", data[data["Platform"] == "GBA"]["Global_Sales"].sum())
```

```python
#計算任天堂運動類遊戲的銷售總量
# c1 = data["Publisher"] == "Nintendo"
# c2 = data["Genre"] == "Sports"
# data[c1 & c2]["Global_Sales"].sum()
d = data[
        (data["Publisher"] == "Nintendo") & (data["Genre"] == "Sports")
        ]["Global_Sales"].sum()

print("任天堂運動類遊戲的銷售總量: ", d)
```

```python
#列出北美地區銷售量最高的前五名遊戲
# 北美地區 data["NA_Sales"]
# 排序北美 data["NA_Sales"].sort()
# .sort_values(by= "", ascending=Fales).head()

d = data.sort_values(by="NA_Sales", ascending=False).head()
pd.DataFrame(d.iloc[0:5]["Name"])
```

```python
#列出北美與歐洲地區銷售量總和最高的前五名遊戲
# 北美地區 data["NA_Sales"]
# 歐洲地區 data["EU_Sales"]
# .sort_values(ascending=False).head()
temp = data["NA_Sales"] + data["EU_Sales"]
index = temp.sort_values(ascending=False).head().index
print(index)
pd.DataFrame(data["Name"][index])
```

```python
#找出總共有多少種不同的平台
s_list = data["Platform"].unique()
s_list
```

```python
#寫程式計算不同平台的總銷售量
# data[data["Platform"] == "Wii"]["Global_Sales"].sum()
for s in s_list:
    print(s, ":", data[data["Platform"] == s]["Global_Sales"].sum())
```

```python
# 使用GroupBy來完成上述的工作
data.groupby("Platform")["Global_Sales"].sum() # 使用groupby一定要運算
```
